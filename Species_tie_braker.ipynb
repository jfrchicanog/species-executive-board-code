This is the code used in the SPECIES AGM to break ties in the voting process of the Executive Board.

```python
import random
import time

def random_permutation(list):
  random.seed(time.time_ns())
  return random.sample(list, len(list))
```

Add here the list of candidates with the same number of votes.

```python
ties = ["Candidate 1", "Candidate 2", "Candidate 3"]
```

Run the cell below to get a random permutation of candidates (run it only once per candidate list).

```python
print(random_permutation(ties))
```
